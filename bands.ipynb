## Masse effective : 

Pour calculer la masse effective nous allons approaximer les bandes à des paraboles. Nous savons qu'un parabole correspond à l'équation suivante :$$y=ax^2+bx+c $$ 
Où $y$ représente l'énergie E en eV et $x$ la distance des vecteurs d'ondes en Å. De plus nous savons que pour trouver la masse effective nous devons résoudre l'équation suivante : $$m^*_e=\frac{\hbar^2}{\frac{\partial^2 E}{\partial x^2}} $$

Pour résoudre cette équation nous allons tout d'abord devoir déterminer les différentes constante de E (a, b et c). Pour ce faire nous utilisons des valeur remarquables comme $Gamma$ qui correspons à un $x=0$. Grâce à celui-ci nous pouvons trouver la valeur de la constance c de l'équation parabolique. 


### Pour la drenière bande de valence : 

Comme dit ci-dessus, nous allons utiliser $x=0$ en $Gamma$ pour trouver la constante c. En effet, dans ce cas $y=c$. Nous avons trouvé ce $y$ plus haut pour pouvoir calculer les dispersions. Nous trouvons :

```python
print('y = c =', ev1)
```

Ensuite nous allons utilisé un des maximum ainsi qu'un aute point pour déterminer nos constantes $a$ et $b$. En effet nous connaissons la valeur du $y$ en ce maximum situé en $Z$ et nous pouvons utiliser les données récoltées avant pour obtenir le $y$ de l'autre point situé en $X|Q$ : 

```python
print('y maximum =', ev6 )
print('y autre point =', ev10)
```

En utilisant les valeur des $x$ autrement dit de $Z$ et $X|Q$ qui correspondent respectivement au maximum et à l'autre point, nous pourrons obtenir deux équations à deux inconnues. 

```python
print('Z =',dZ)
print('X|Q =',dXQ)
```

En arrondissant ces valeurs, nous obtenons les deux équations suivantes : 
$$ 2,626^2a + 2,626b - 0,733 = 0$$
$$ 4,317^2a + 4,317b - 0,733 = -0,185$$
D'où nous pouvons trouver $a = -0,089 $ et $b = 0,51$
Nous abtenons finalement l'équation de E :
$$ E = -0,089x^2 + 0,51b - 0,733 $$
Grâce à cette dernière nous calculons la dérivée seconde de E qui est égale à $2a$ soit une constante = $-0,178$. Cependant avant de trouver $m^*_e$ il faut se pencher sur les unités de toutes ces données. En effet $\hbar$ est en $J.s$ . Il faut donc que notre constante $a$ qui a pour l'instant comme unité des $\frac{eV}{Å^2}$ devienne des $\frac{J}{m}$. Nous devons donc faire : $\frac{-0,178*1,602181*10^{-19}}{10^{-10^2}} = -2,852 \frac{J}{m} $ 
Dés lors en reprenant l'équation de la masse effective nous pouvons calculer celle-ci : 
$$ m^*_e = \frac{(6,626 10^{-34})^2}{-2,852} = -1,539 10^{-67} kg$$
