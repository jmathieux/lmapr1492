## Dispersions :

Par définition, la dispersion est une variation d’énergie en fonction d'un espace. Dans notre cas ça tombe bien car notre axe des ordonnées nous donne une énergie et notre axe des abscises un espace. Nous allons donc trouver pour chaque espace sa taille ainsi que le delta d'énergie qui y correspond. Grâce à cela nous pourrons calculer les pentes et ainsi trouver les directions dans lesquelles il y a le plus de dispersion (correspondant à la pente la plus forte) et le moins de dispersion (correspondant à la pente la plus faible). 

### Données pour la bande de valence (partie sous la bande interdite indirecte) :

```python
t=[["Gamma-L  :", ev1, ev2, ev2-ev1, d1, (ev2-ev1)/d1], 
       ["L-B1 :", ev3, ev4, ev4-ev3, d2, (ev4-ev3)/d2],
       ["B-Z  :", ev5, ev6,ev6-ev5, d3, (ev6-ev5)/d3],
       ["Z-Gamma :", ev7, ev8, ev8-ev7, d4, (ev8-ev7)/d4],
       ["Gamma-X :", ev9, ev10, ev10-ev9, d5, (ev10-ev9)/d5],
       ["Q-F  :", ev11, ev12, ev12-ev11, d6, (ev12-ev11)/d6],
       ["F-P1  :", ev13, ev14, ev14-ev13, d7, (ev14-ev13)/d7],
       ["P1-Z :", ev15, ev16, ev16-ev15, d8, (ev16-ev15)/d8],
       ["L-P  :", ev17, ev18, ev18-ev17, d9, (ev18-ev17)/d9],
]
print(tabulate(t, headers =[" ","Energies début (eV)","Energies fin (eV)","Deltas energies (eV)","Deltas distances (Å)","Pentes"], 
               stralign="center",numalign="center"),"\n")
```

### Données pour la bande de conduction (partie au dessus la bande interdite indirecte) :

```python
t=[["Gamma-L  :", ec1, ec2, ec2-ec1, d1, (ec2-ec1)/d1],
       ["L-B1 :", ec3, ec4, ec4-ec3, d2, (ec4-ec3)/d2],
       ["B-Z  :", ec5, ec6, ec6-ec5, d3, (ec6-ec5)/d3],
       ["Z-Gamma :", ec7, ec8, ec8-ec7, d4, (ec8-ec7)/d4],
       ["Gamma-X :", ec9, ec10,ec10-ec9, d5, (ec10-ec9)/d5],
       ["Q-F  :", ec11, ec12, ec12-ec11, d6, (ec12-ec11)/d6],
       ["F-P1  :", ec13, ec14,ec14-ec13, d7, (ec14-ec13)/d7],
       ["P1-Z :", ec15, ec16, ec16-ec15, d8, (ec16-ec15)/d8],
       ["L-P  :", ec17, ec18,ec18-ec17, d9, (ec18-ec17)/d9],
]
print(tabulate(t, headers =[" ","Energies début (eV)","Energies fin (eV)","Deltas energies (eV)","Deltas distances (Å)","Pentes"], 
               stralign="center",numalign="center"),"\n")
```

Grâce à ces tableaux de données, nous pouvons voir que pour la bande de valence, le plus de dispersion se trouve dans la direction $Z$ $Gamma$ et le moins de dispersion se trouve dans la direction $F$ $P_1$. Pour la bande de conduction, le plus de dispersion se trouve dan sla direction $Z$ $Gamma$ et le moins de dispersion se trouve dans la direction $Gamma$ $L$. Ces direction sont indiquées sur la photo suivante à l'aide de flèches. En rouge les directions pour la bande de conduction et en vert les directions pour la bande de valence. Les "+" désignent les directions avec le plus de dispersion et les "-" designent les directions avec le moins de dispersion.
