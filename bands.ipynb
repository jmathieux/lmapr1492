## Dispersions :

Par définition, la dispersion est une variation d’énergie en fonction d'un espace. Dans notre cas ça tombe bien car notre axe des ordonnées nous donne une énergie et notre axe des abscises un espace. Nous allons donc trouver pour chaque espace sa taille ainsi que le delta d'énergie qui y correspond. Grâce à cela nous pourrons calculer les pentes et ainsi trouver les directions dans lesquelles il y a le plus de dispersion (correspondant à la pente la plus forte) et le moins de dispersion (correspondant à la pente la plus faible). 

### Données pour la bande de valence  :

```python
t=[["Gamma-L  :", ev1, ev2, ev2-ev1, d1, (ev2-ev1)/d1], 
       ["L-B1|B :", ev3, ev4, ev4-ev3, d2, (ev4-ev3)/d2],
       ["B1|B-Z  :", ev5, ev6,ev6-ev5, d3, (ev6-ev5)/d3],
       ["Z-Gamma :", ev7, ev8, ev8-ev7, d4, (ev8-ev7)/d4],
       ["Gamma-X|Q :", ev9, ev10, ev10-ev9, d5, (ev10-ev9)/d5],
       ["X|Q-F  :", ev11, ev12, ev12-ev11, d6, (ev12-ev11)/d6],
       ["F-P1  :", ev13, ev14, ev14-ev13, d7, (ev14-ev13)/d7],
       ["P1-Z|L :", ev15, ev16, ev16-ev15, d8, (ev16-ev15)/d8],
       ["Z|L-P  :", ev17, ev18, ev18-ev17, d9, (ev18-ev17)/d9],
]
print(tabulate(t, headers =[" ","Energies début","Energies fin","Deltas energies","Distances","Pentes"], 
               stralign="center",numalign="center"),"\n")
print('Les énergies sont données en eV et les distances en Å^-1')
```

## Masse effective : 

Pour calculer la masse effective nous allons approaximer les bandes à des paraboles. Nous savons qu'un parabole correspond à l'équation suivante :$$y=ax^2+bx+c $$ 
Où $y$ représente l'énergie E en eV et $x$ la distance des vecteurs d'ondes en Å. De plus nous savons que pour trouver la masse effective nous devons résoudre l'équation suivante : $$m^*_e=\frac{\hbar^2}{\frac{\partial^2 E}{\partial x^2}} $$

Pour résoudre cette équation nous allons tout d'abord devoir déterminer les différentes constante de l'équation de E (a, b et c). Pour ce faire nous utilisons allons utiliser 3 points. Pour la bande de valence, nous allons prendre un point avant le sommet situé en $Z$, le sommet ainsi que le point après ce sommet. Pour la bande de conduction, nous allons prendre un point avant le minimum situé en $L$, le minimum ainsi que le point après ce minimum. 


### Pour la dernière bande de valence : 

Comme dit ci-dessus, nous allons utiliser les coordonnées des points que voici : (absice en $eV$ et ordonnée en $Å^{-1}$)

```python
print('Pour les abscisses du point avant le sommet Z, de le sommet Z et du point après le sommet Z (respectivement): ')
print(vect_xv,'\n')
print('Pour les ordonnées du point avant le sommet Z, de le sommet Z et du point après le sommet Z (respectivement): ')
print(vect_yv,'\n')
```

Ensuite, grace à une commande python, nous allons calculer nos indices a, b et c afin d'obtenir notre équation de E. 

```python
print('[a,b,c]=', indices_v)
```

Nous pouvons, dès lors, écrire l'équation de notre parabole approximative de la disperson telle que : $$ E = -2,766x^2 + 14,232x - 18,295 $$
Il est maintenant judicieux de s'intéresser aux unités. En effet, pour l'instant nous avons des énergies en $eV$ et des distances en $Å^{-1}$. Nous voulons obtenir une masse effective en $kg$. Nous allons alors changer nos $eV$ en $J$ et nos $Å^{-1}$ en $m^{-1}$. Puisque la dérivée seconde d'une équation du second degré est égale à $2a$, nous allons effectuer ce changement d'unités uniquement pour la constante $a$ de l'équation.

```python
a_v=indices_v[0]
eV_J=sc.physical_constants['electron volt-joule relationship'][0]
new_a=a_v*eV_J*sc.angstrom**2
print('Constante a en [J.m^2] =', new_a)
```

On obtient donc une constante $$a = -2,766 [{eV}.{Å^2}] = -4,432*10^{-39} [J.m^2]$$

En reprenant notre équation de la massse effective, notre changement d'unités et en sachant que la dérivée seconde d'une équation du second degré est égale à $2a$ nous pouvons déterminer notre mase effective au sommet de la dernière bande de valance. 


```python
me_v=sc.hbar**2/(2*new_a)
me_ve=me_v/sc.electron_mass
print('En sachant que la constante de Planck =',sc.hbar,'[J.s] et que la masse d un électron =',sc.electron_mass,'[kg] on trouve :')
print('m*_e =',me_v,'kg')
print('m*_e =',me_ve,'m_el')
```

En effet, $$ m^*_e = \frac{\hbar^2}{2a} = -1,254*10^{-30} [kg] $$
Mais aussi $$m^*_e = -1,377 m_{electron}$$

Il est maintenant judicieux de s'intéresser aux unités. En effet, pour l'instant nous avons des énergies en $eV$ et des distances en $Å^{-1}$. Nous voulons obtenir une masse effective en $kg$. Nous allons alors changer nos $eV$ en $J$ et nos $Å^{-1}$ en $m^{-1}$.
