# Code : 

```python
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen.ext.matproj import MPRester
with MPRester("9pWDrcCtfsNndSHF") as m:
    structure = m.get_structure_by_material_id("mp-13694")
    
    cal1 = XRDCalculator() #avec symprec à 0.0 (par défaut) on obtient les pics avec les indices hkl 
    xrd1 = cal1.get_pattern(structure)
    
    cal2 = XRDCalculator(symprec=0.1) #avec symprec à 0.1 on obtient les mêmes valeurs que sur Materials Project
    xrd2 = cal2.get_pattern(structure)
    
#Obtention des pics 
    pics1 = xrd1.hkls
    pics2 = xrd2.hkls
```

# Résultats : 

## Affichage du graphe de diffraction 

```python
print(cal1.show_plot(structure, annotate_peaks=False))
```
